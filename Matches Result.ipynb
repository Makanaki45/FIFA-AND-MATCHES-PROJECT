```python
#loading the required libraries
import pandas as pd
from matplotlib import pyplot as pt
import seaborn as sns
```

```python
#loading the ipl matches dataset
ipl = pd.read_csv('matches.csv')
```

```python
#having a glance at the first five records of the dataset
ipl.head()
```

```python
#Lookin at the number of rows and columns in the dataset
ipl.shape
```

```python
#Looking at the datatype of the columns
ipl.info()
```

```python
#looking at the statistical aspect of the data
ipl.describe()
```

```python
#Getting the frequency of most man of the match awards
ipl['player_of_match'].value_counts()
```

```python
#Getting the top 10 players with most man of the match awards
ipl['player_of_match'].value_counts()[0:10]
```

```python
#Getting the top 10 players with most man of the match awards
ipl['player_of_match'].value_counts()[0:5]
```

```python
#making a bar-plot for the top 5 players with most man of the match awards
pt.figure(figsize=(10,5))
pt.bar(list(ipl['player_of_match'].value_counts()[0:5].keys()), list(ipl['player_of_match'].value_counts()[0:5]), color=['Green','Red','Orange', 'Green', 'Blue'])
pt.show()
```

```python
#Getting the frequency of result column
ipl['result'].value_counts()
```

```python
#Finding out the number of toss wins w.r.t each team
ipl['toss_winner'].value_counts()
```

```python
#Extracting the records where a team won batting first
battling_run = ipl[ipl['win_by_runs']!=0]
```

```python
#looking at the head
battling_run
```

```python
#Making a histogram 
pt.figure(figsize=(7,7))
pt.hist(battling_run['win_by_runs'])
pt.show()
```

```python
#Finding out the number of wins w.r.t each team after battling run
battling_run['winner'].value_counts()
```

```python
#Making a bar-plot for top 3 teams with most wins after battling run
pt.figure(figsize=(7,7))
pt.bar(list(battling_run['winner'].value_counts()[0:3].keys()),list(battling_run['winner'].value_counts()[0:3]),color=["blue","yellow","orange"])
pt.show()
```

```python
#Making a pie chart
pt.figure(figsize=(20,10))
pt.pie(list(battling_run['winner'].value_counts()),labels=list(battling_run['winner'].value_counts().keys()),autopct='%0.1f%%')
pt.show()
```

```python
#extracting those records where a team has won after battling wickets
battling_wickets =ipl[ipl['win_by_wickets']!=0]
```

```python
#looking at the head
battling_wickets.head()
```

```python
#Making a histogram for frequency of wins w.r.t number of wickets
pt.figure(figsize=(7,7))
pt.hist(battling_wickets['win_by_wickets'],bins=30)
pt.show()
```

```python
#Finding out the frequency of number of wins w.r.t each time after battling wickets
battling_wickets['winner'].value_counts()
```

```python
#Making a bar plot for top-3 teams with most wins after battling wickets
pt.figure(figsize=(7,7))
pt.bar(list(battling_wickets['winner'].value_counts()[0:3].keys()),list(battling_wickets['winner'].value_counts()[0:3]),color=["blue","green","orange"])
pt.show()
```

```python
#Making a pie chart for distribution of most wins after battling wickets
pt.figure(figsize=(7,7))
pt.pie(list(battling_wickets['winner'].value_counts()),labels=list(battling_wickets['winner'].value_counts().keys()),autopct='%0.1f%%')
pt.show()
```
